```python
deliveries_df.isnull().sum()
deliveries_df_copy = deliveries_df.copy()
deliveries_df_copy.drop(columns=['day', 'month'], axis=1, inplace=True)
```

```python
deliveries_df_copy['ball'] = pd.to_numeric(deliveries_df_copy['ball'], errors='coerce').round(1)
print(deliveries_df_copy['ball'].isnull().sum())
```

```python
# Calculate over_ball and store in a variable
deliveries_df_copy['over_completed'] = deliveries_df_copy['ball'].astype(int)
deliveries_df_copy['valid_ball'] = (deliveries_df_copy['wides'] == 0) & (deliveries_df_copy['noballs'] == 0)

deliveries_df_copy['over_ball'] = deliveries_df_copy.groupby(['match_number', 'innings','over_completed'])['valid_ball'].cumsum()
deliveries_df_copy.head(20)
#deliveries_df_copy.drop(columns=['valid_ball'], axis=1, inplace=True)
deliveries_df_copy[deliveries_df_copy['valid_ball']==False].head()
```

```python
deliveries_df_copy.head(5)
```

```python
deliveries_df_copy['total_runs'] = deliveries_df_copy.groupby(['match_number', 'innings'])[['runs_off_bat', 'extras']].cumsum().sum(axis=1) #row wise addition
deliveries_df_copy[['runs_off_bat', 'extras', 'total_runs']].tail(50)
```

```python
deliveries_df_copy[['noballs', 'wides', 'total_runs','over_ball','valid_ball']].tail(50)
```

```python
deliveries_df_copy['is_wicket'] = deliveries_df_copy['wicket_type'].notna().astype(int)
deliveries_df_copy[['wicket_type', 'is_wicket']].head(50)
```

```python
deliveries_df_copy['total_wickets'] = deliveries_df_copy.groupby(['match_number', 'innings'])['is_wicket'].cumsum()
deliveries_df_copy.drop(columns=['is_wicket'], axis=1, inplace=True)
deliveries_df_copy.drop(columns=['valid_ball','ball'], axis=1, inplace=True)
deliveries_df_copy.tail(20)
```

```python
deliveries_df_copy['run_rate']=(deliveries_df_copy['total_runs']/(deliveries_df_copy['over_ball']+deliveries_df_copy['over_completed']*6)*6).round(2)
deliveries_df_copy.tail()
```

```python
deliveries_df_copy['balls_left'] = 120 - (deliveries_df_copy['over_ball'] + deliveries_df_copy['over_completed'] * 6)
deliveries_df_copy.tail()
```

```python
#  Calculate total runs in the first innings for each match
first_innings_target = deliveries_df_copy[deliveries_df_copy['innings'] == 1].groupby('match_number')['total_runs'].max() + 1
second_innings_total = deliveries_df_copy[deliveries_df_copy['innings'] == 2].groupby('match_number')['total_runs'].max()

#  Assign the target 
deliveries_df_copy['target'] = deliveries_df_copy['match_number'].map(first_innings_target)
deliveries_df_copy['second_innings_total'] = deliveries_df_copy['match_number'].map(second_innings_total)
deliveries_df_copy.head()
```

```python
def calculate_req_runs(total_runs,target,innings):
    if innings == 1:
        return 0
    else:
        return target - total_runs
```

```python
deliveries_df_copy['req_runs'] = deliveries_df_copy.apply(lambda x: calculate_req_runs(x['total_runs'],x['target'],x['innings']),axis=1)
deliveries_df_copy.tail()
deliveries_df_copy.head()
```

```python
def calculate_req_runrate(req_runs,innings,over_ball,over_completed):
    if innings == 1:
        return 0
    elif (over_ball + over_completed*6) == 120:
        return 0
    else:
        return ((req_runs)/(120 - (over_ball + over_completed*6))*6)
```

```python
deliveries_df_copy['req_runrate'] = deliveries_df_copy.apply(lambda x: calculate_req_runrate(x['req_runs'],x['innings'],x['over_ball'],x['over_completed']),axis=1).round(2)
deliveries_df_copy.head()
deliveries_df_copy.tail(5)
```

```python
venue_1st_innings_avg = deliveries_df_copy[deliveries_df_copy['innings']==1].groupby(['venue','innings'])['target'].unique().apply(lambda x: (x-1).mean()).round(0).astype(int)

venue_1st_innings_avg 
```

```python
venue_2nd_innings_avg = deliveries_df_copy[deliveries_df_copy['innings']==2].groupby(['venue','innings'])['second_innings_total'].unique().apply(lambda x: x.mean()).round(0).astype(int)
venue_2nd_innings_avg
```

```python
venue_dataframe = pd.concat([venue_1st_innings_avg,venue_2nd_innings_avg],axis=1).reset_index()
venue_dataframe.rename(columns={'second_innings_total': 'venue_avg_1','target': 'venue_avg_2'},inplace=True)
venue_dataframe.fillna(0,inplace=True)
venue_dataframe['venue_avg'] = (venue_dataframe['venue_avg_1'] + venue_dataframe['venue_avg_2'])
venue_dataframe.drop(columns=['venue_avg_1','venue_avg_2'],axis=1,inplace=True)
venue_dataframe
```

```python
deliveries_df_copy['venue_avg'] = pd.merge(deliveries_df_copy,venue_dataframe,how='inner', on=['venue','innings'])['venue_avg']
```
